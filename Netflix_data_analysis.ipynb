```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=pd.read_csv('mymoviedb.csv',lineterminator='\n')
x.head()
```

```python
# Frequent genre of movies released on Netflix.
u=pd.pivot_table(data=x,index='Genre',values='Title',aggfunc='count').sort_values(by='Title',ascending=False).head(20)
u=u.head(5).reset_index()

# visualizationa
color=['#f1c40f','#f39c12','#e67e22','#fdcb6e','#f0932b']
plt.bar(u['Genre'],u['Title'],color=color)
plt.xticks(rotation=40)
plt.title('Frequent genre of movies released')
plt.ylabel
plt.xlabel('Genre',size=11)


plt.show()
```

```python
# Genre with highest Vote.
D=pd.pivot_table(data=x,index='Genre',values='Vote_Count',aggfunc='sum').sort_values(by='Vote_Count',ascending=False)
D=D.reset_index().head()

# Visualization
explode=[0.1 if genre =="Drama" else 0 for genre in D['Genre']]
color=['#ffcccc','#ffb8b8','#f8a5c2','#f78fb3','#ff6b81']
plt.bar(D['Genre'],D['Vote_Count'],color=color,label=D["Genre"])
plt.xticks(rotation=90)
plt.title('Genre with highest Vote')
plt.legend()
plt.show()
```

```python
# Movie and Genre with highest Popularity.
h=pd.pivot_table(data=x,index=['Title','Genre'],values='Popularity',aggfunc='max').sort_values(by='Popularity',ascending=False)
h.head().reset_index
```

```python
# Movie and Genre with Lowest Popularity.
H=pd.pivot_table(data=x,index=['Title','Genre'],values='Popularity',aggfunc='min').sort_values(by='Popularity',ascending=True)
H.head().reset_index
```

```python
x['Release_Date']=pd.to_datetime(x['Release_Date'])
```

```python
x['Release_Date']=x['Release_Date'].dt.year
```

```python
# Year with most filmed movies.
df=pd.pivot_table(data=x,index='Release_Date',values='Title',aggfunc='count').sort_values(by='Title',ascending=False)
df=df.head().reset_index()

# Visualization
explode=[0.1 if year ==2021 else 0 for year in df['Release_Date']]
color=['#82ccdd','#60a3bc','#38ada9','#079992','#3c6382']
plt.pie(df['Title'],labels=df['Release_Date'],autopct="%0.1f%%",explode=explode,colors=color)
plt.title('Year with most filmed movies')
plt.show()
```

>

The year 2021 accounts for the highest percentage of movies filmed.
